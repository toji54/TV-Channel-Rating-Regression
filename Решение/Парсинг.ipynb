### Извлечение дополнительной информации

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = 'https://tv.mail.ru'

def parse_program_details(relative_url):
    try:
        full_url = base_url + relative_url
        response = requests.get(full_url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        year_release = soup.find(class_='nowrap').text
        age_limit = soup.find(class_='label label_restrict label_restrict_small valign_middle').text
        presenters_div = soup.find(class_='table__cell padding_bottom_10')
        presenters_links = presenters_div.find_all('a')
        presenters = [link.get_text() for link in presenters_links]

        link = soup.find('a', class_='link link_inline link_underline-line')['href']
        response = requests.get(link)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        description = soup.find(class_='text text_inline text_light_medium text_fixed valign_baseline p-movie-info__description-text').text
        rating = soup.find(class_='text text_bold_huge text_fixed').text
        number_ratings = soup.find(class_='text text_light_medium text_fixed margin_left_10').text
        categories = [item.text for item in soup.find_all(class_='badge badge_gray badge_gray_rgba badge_border_off badge_link')]

        return {
            'Year Release': year_release,
            'Description': description,
            'Age Limit': age_limit,
            'Rating': rating,
            'Number of Ratings': number_ratings,
            'Categories': categories,
            'Presenters': presenters
        }
    except Exception as e:
        print(f"Error parsing program details for {relative_url}: {e}")
        return None

# Пример использования
channels = soup.find_all(class_='p-channels__item__info__title')
channel_programs = []

for channel in channels:
    channel_name = channel.text

    if channel_name.endswith('live'):
        channel_name = channel_name[:-4].strip()

    if channel_name == 'РБК':
      break

    channel_parent = channel.find_parent(class_='p-channels__item')
    program_items = channel_parent.find_all(class_='p-programms__item')

    for program in program_items:
        program_link = program.find('a', class_='p-programms__item__inner')
        href = program_link['href']
        time = program.find(class_='p-programms__item__time').text.strip()
        full_url = base_url + href

        response = requests.get(full_url)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        name = soup.find(class_='hdr__text js-event-popup__title').text
        print(time, channel_name, name)
        existing_program = next((d for d in channel_programs if d['Channel'] == channel_name and d['Program'] == name), None)

        if existing_program:
            existing_program['Time'].append(time)
        else:
            additional_data = parse_program_details(href)
            if additional_data:
                channel_programs.append({
                    'Channel': channel_name,
                    'Time': [time],
                    'Program': name,
                    **additional_data
                })

df = pd.DataFrame(channel_programs)
df
```

```python
df.to_csv(f'{title}', index=False, encoding='utf-8')
```
